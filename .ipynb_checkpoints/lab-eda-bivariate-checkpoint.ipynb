```python
#Crosstab Analysis:
#Create a crosstab between the product category and the isBestSeller status.

crosstab = pd.crosstab(df['category'], df['isBestSeller'])

# Display the crosstab
crosstab
```

```python
#Are there categories where being a best-seller is more prevalent?

best_seller_proportion = df.groupby('category')['isBestSeller'].mean().sort_values(ascending=False)
best_seller_proportion
```

```python
#Statistical Tests:

#Conduct a Chi-square test to determine if the best-seller distribution is independent of the product category
crosstab = pd.crosstab(df['category'], df['isBestSeller'])
print(crosstab)
from scipy.stats import chi2_contingency
chi2, p, _, _ = chi2_contingency(crosstab)

# Print the results
print(f"Chi-Square Statistic: {chi2}")
print(f"p-value: {p}")
```

Since p_value > 0.05 , there is a relationship between the two categorical values.

```python
#Compute Cramér's V to understand the strength of association between best-seller status and category.

from scipy.stats.contingency import association

# Computing the association between variables in 'crosstab' using the "cramer" method
association(crosstab, method="cramer")
```

0.1 represents a slight association between the two categorical variables.

```python
import seaborn as sns
import matplotlib.pyplot as plt
crosstab.plot(kind="bar", stacked=True)
plt.ylim(0,20250)
```

```python
# Calculate Q1 (25th percentile) and Q3 (75th percentile) for the 'price' column
Q1 = df['price'].quantile(0.25)
Q3 = df['price'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

df_no_outliers = df[(df['price'] >= lower_bound) & (df['price'] <= upper_bound)]

# Check the shape of the dataframe before and after removing outliers
print(f"Original dataset size: {df.shape}")
print(f"Dataset size after removing outliers: {df_no_outliers.shape}")
```

```python
#Use a violin plot to visualize the distribution of price across different product categories. Filter out the top 20 categories based on count for better visualization.
import seaborn as sns
import matplotlib.pyplot as plt

top_20_categories = df_no_outliers['category'].value_counts().nlargest(20).index
df_top_20 = df_no_outliers[df_no_outliers['category'].isin(top_20_categories)]


plt.figure(figsize=(12, 8))
sns.violinplot(x='category', y='price', data=df_top_20, palette='Set3')
plt.xticks(rotation=90)
plt.title('Price Distribution Across Product Categories (Top 20 Categories)')
plt.xlabel('Product Category')
plt.ylabel('Price')
plt.show()
```

```python
#Which product category tends to have the highest median price? Don't filter here by top categories.
Frangrances and Men have the highest medain according to the violin graph
```

```python
#Create a bar chart comparing the average price of products for the top 10 product categories (based on count).
# Getting the top 10 categories
top_10_categories = df_no_outliers['category'].value_counts().nlargest(10).index

# Filter the dataset for these top 10 categories
df_top_10 = df_no_outliers[df_no_outliers['category'].isin(top_10_categories)]

# Group by category and calculate the average price
avg_price_by_category = df_top_10.groupby('category')['price'].mean().sort_values(ascending=False)

# Plot the bar chart
plt.figure(figsize=(10, 6))
sns.barplot(x=avg_price_by_category.index, y=avg_price_by_category.values, palette='Blues_d')

# Add title and labels
plt.title('Average Price of Products by Top 10 Categories')
plt.xlabel('Product Category')
plt.ylabel('Average Price')
plt.xticks(rotation=90)
plt.show()
```

#Which product category commands the highest average price? Don't filter here by top categories.
Sports and Outdoor has the highest average price

```python
#Visualize the distribution of product ratings based on their category using side-by-side box plots. Filter out the top 10 categories based on count for better visualization.
# Create a box plot for product ratings across categories
plt.figure(figsize=(12, 8))
sns.boxplot(x='category', y='stars', data=df_top_10, palette='Set2')

# Rotate x-axis labels for better readability
plt.xticks(rotation=90)
plt.title('Distribution of Product Ratings by Category (Top 10 Categories)')
plt.xlabel('Product Category')
plt.ylabel('Rating')
plt.show()
```

#Which category tends to receive the highest median rating from customers? Don't filter here by top categories.

Skincare has the highest median rating from customers
